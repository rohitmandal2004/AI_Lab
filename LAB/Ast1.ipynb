```python
import pandas as pd

car_description = {
    "Parameter": ["Dzire", "City", "Eco", "Amaze", "Indica", "Nexon", "WagonR"],
    "CAGR in %": [52, 8, 56, 45, 58, 78, 80],
    "Unit Sold": [12560, 234, 2000, 2800, 3400, 7000, 12000],
    "Benefit": [5000, 15000, 20000, 25000, 12000, 25000, 35000],
    "Sale in WB": [2000, 20, 3000, 2500, 5000, 12000, 11500]
}

df1 = pd.DataFrame(car_description)

print(df1)

print("\nMaximum Values")
print(df1.max())

print("\nMinimum Values")
print(df1.min())

print("\nStandard Deviation of Unit Sold")
print(df1["Unit Sold"].std())
```
